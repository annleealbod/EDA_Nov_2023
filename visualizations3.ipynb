# The Fatherlode

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("the fatherlode.csv")
df.head()
```

# Correlation Heatmaps First Shift

```python
# Select relevant columns (excluding year_quarter_comp_start)
selected_columns = ['head_count_requested_fs', 'fill_rate_fs', 'shift_lead_time_fs', 'total_head_count_minus_fs']
df_selected = df[selected_columns]

# Filter data for each year_quarter_comp_start
df_2022Q3 = df[df['year_quarter_comp_start'] == '2022Q3'][selected_columns]
df_2022Q4 = df[df['year_quarter_comp_start'] == '2022Q4'][selected_columns]
df_2023Q1 = df[df['year_quarter_comp_start'] == '2023Q1'][selected_columns]
df_2023Q2 = df[df['year_quarter_comp_start'] == '2023Q2'][selected_columns]

# Convert numeric columns to float (optional, if not already)
numeric_columns = df_selected.columns[df_selected.dtypes.apply(lambda x: np.issubdtype(x, np.number))]
df_selected[numeric_columns] = df_selected[numeric_columns].astype(float)

# Create a correlation matrix for the entire dataset
corr_all = df_selected.corr()

# Create a correlation matrix for each category
corr_2022Q3 = df_2022Q3.corr()
corr_2022Q4 = df_2022Q4.corr()
corr_2023Q1 = df_2023Q1.corr()
corr_2023Q2 = df_2023Q2.corr()

# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Set an overall title for the entire grid
fig.suptitle('First Shift Effect by Company Start Quarter', fontsize=16)

# Plot the correlation heatmaps for each category
sns.heatmap(corr_2022Q3, annot=True, cmap='coolwarm', ax=axes[0, 0])
axes[0, 0].set_title('2022Q3')

sns.heatmap(corr_2022Q4, annot=True, cmap='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title('2022Q4')

sns.heatmap(corr_2023Q1, annot=True, cmap='coolwarm', ax=axes[1, 0])
axes[1, 0].set_title('2023Q1')

sns.heatmap(corr_2023Q2, annot=True, cmap='coolwarm', ax=axes[1, 1])
axes[1, 1].set_title('2023Q2')

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Add space for the overall title

# Show the plot
plt.show()

```
